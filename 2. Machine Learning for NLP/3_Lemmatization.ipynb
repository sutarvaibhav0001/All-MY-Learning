### ğŸ”¹ **Definition**

Lemmatization is the process of **reducing a word to its base or dictionary form**, known as the **lemma**.

### ğŸ”¹ **How Lemmatization Works**

1. **Tokenization:** Split text into words.
2. **POS Tagging:** Identify each wordâ€™s part of speech (noun, verb, adjective, etc.).
3. **Dictionary Lookup:** Use a lexical database (like *WordNet*) to find the base form (lemma) according to POS.

### ğŸ”¹ **Example Using Python (NLTK)**

```python
# Example: Lemmatization in Python using NLTK

import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download resources (run once)
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()

# Helper function to get POS tag
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        'J': wordnet.ADJ,
        'N': wordnet.NOUN,
        'V': wordnet.VERB,
        'R': wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

words = ["running", "ate", "better", "rocks", "cars"]

for word in words:
    lemma = lemmatizer.lemmatize(word, get_wordnet_pos(word))
    print(f"{word:10} â†’ {lemma}")
```

**Output:**

```
running    â†’ run
ate        â†’ eat
better     â†’ good
rocks      â†’ rock
cars       â†’ car
```
