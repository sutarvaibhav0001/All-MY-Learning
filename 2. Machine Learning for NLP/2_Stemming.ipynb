## ğŸ”¹ **Example in Python (NLTK)**

```python
# Example: Stemming in Python using NLTK

from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

# Initialize stemmers
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")

words = ["running", "runs", "easily", "fairly", "connection", "connected"]

print("Word | Porter | Lancaster | Snowball")
for w in words:
    print(f"{w:10} | {porter.stem(w):10} | {lancaster.stem(w):10} | {snowball.stem(w):10}")
```

| ---------- | ------- | --------- | -------- |
| running    | run     | run       | run      |
| runs       | run     | run       | run      |
| easily     | easili  | easy      | easili   |
| fairly     | fairli  | fair      | fair     |
| connection | connect | connect   | connect  |
| connected  | connect | connect   | connect  |

## ğŸ”¹ **When to Use Stemming**
