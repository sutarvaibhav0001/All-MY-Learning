## Steps to Push a Folder to GitHub

#### first do this imp step - go to github and then 

### 1. Go to Home tab


### 2. Then click on New (green tab)


### 3. Create new repo by giving folder name 


---

1. Open Terminal / Command Prompt

Navigate to your project folder:
cd /path/to/your/folder

---

2. Initialize Git (if not already)

git init

---

3. Add a Remote Repository

Replace the URL with your GitHub repo URL:

git remote add origin https://github.com/username/repo-name.git

example - git remote add origin https://github.com/sutarvaibhav0001/Complete-Python-Bootcamp.git

---

4. Add Files to Git

git add .

(The . adds everything in the current folder.)

---

5. Commit the Changes

git commit -m "Initial commit"

---

6. Push to GitHub

If it's a new repo, push to the main branch (or master depending on your setup):

example - git remote add origin https://github.com/sutarvaibhav0001/Complete-Python-Bootcamp.git   (dont forget to add .git at end)

git branch -M main

git push -u origin main
