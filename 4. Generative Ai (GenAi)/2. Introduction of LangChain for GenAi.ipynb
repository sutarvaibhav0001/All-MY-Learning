1. LangChain is an open-source framework designed to simplify the creation of applications using large language models (LLMs).

2. It helps AI developers connect LLMs such as GPT-4 with external data and computation. 

3. This framework comes for both Python and JavaScript.

### Key Components of LangChain



1. Chains: Chains define sequences of actions, where each step can involve querying an LLM, manipulating data or interacting with external tools. There are two types:
* Simple Chains: A single LLM invocation.
Multi-step Chains: Multiple LLMs or actions combined, where each step can take the output from the previous step.

2. Prompt Management: LangChain facilitates managing and customizing prompts passed to the LLM. Developers can use PromptTemplates to define how inputs and outputs are formatted before being passed to the model. It also simplifies tasks like handling dynamic variables and prompt engineering, making it easier to control the LLM's behavior.

3. Agents: Agents are autonomous systems within LangChain that take actions based on input data. They can call external APIs or query databases dynamically, making decisions based on the situation. These agents leverage LLMs for decision-making, allowing them to respond intelligently to changing input.

4. Vector Database: LangChain integrates with a vector database which is used to store and search high-dimensional vector representations of data. This is important for performing similarity searches, where the LLM converts a query into a vector and compares it against the vectors in the database to retrieve relevant information.
* Vector database plays a key role in tasks like document retrieval, knowledge base integration or context-based search providing the model   with dynamic, real-time data to enhance responses.

5. Models: LangChain is model-agnostic meaning it can integrate with different LLMs such as OpenAI's GPT, Hugging Face models, DeepSeek R1 and more. This flexibility allows developers to choose the best model for their use case while benefiting from LangChains architecture.

6. Memory Management: LangChain supports memory management allowing the LLM to "remember" context from previous interactions. This is especially useful for creating conversational agents that need context across multiple inputs. The memory allows the model to handle sequential conversations, keeping track of prior exchanges to ensure the system responds appropriately.

# 5 Basic LLM Example

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")

response = llm.invoke("Explain Transformers in 2 lines")
print(response.content)
```

# 6 Prompt Templates

Instead of hardcoding prompts, we use templates.

```python
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms."
)

prompt = template.format(topic="Vector Databases")
print(prompt)
```

---

# 7 Chains in LangChain

Chain = Combine multiple components together.

### Example: LLM Chain

```python
from langchain.chains import LLMChain
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate

llm = ChatOpenAI(model="gpt-4o-mini")

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in 3 lines"
)

chain = LLMChain(llm=llm, prompt=prompt)

response = chain.run("Embeddings")
print(response)
```

---

# 8 Sequential Chain

Output of one chain  input of another chain.

```python
from langchain.chains import SimpleSequentialChain

chain1 = LLMChain(llm=llm, prompt=PromptTemplate(
    input_variables=["topic"],
    template="Write short note on {topic}"
))

chain2 = LLMChain(llm=llm, prompt=PromptTemplate(
    input_variables=["text"],
    template="Summarize this:\n{text}"
))

overall_chain = SimpleSequentialChain(chains=[chain1, chain2])

result = overall_chain.run("Transformers")
print(result)
```

---

# 9 Memory in LangChain

Memory stores conversation history.

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory
)

conversation.run("Hi, I am Vaibhav")
conversation.run("What is my name?")
```

Memory types:

* Buffer Memory
* Window Memory
* Summary Memory
* Vector Memory

#  Document Loaders

Used to load external documents.

```python
from langchain.document_loaders import TextLoader

loader = TextLoader("data.txt")
documents = loader.load()
```

Supported:

* PDF
* CSV
* Web pages
* Notion
* YouTube transcripts

---

# 11 Text Splitters

Split large documents into chunks.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

docs = splitter.split_documents(documents)
```

---

# 12 Embeddings

Convert text  vector representation.

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

vector = embeddings.embed_query("What is AI?")
print(len(vector))
```

---

# 13 Vector Stores

Store embeddings for similarity search.

Popular vector DBs:

* FAISS
* Chroma

Example with FAISS:

```python
from langchain.vectorstores import FAISS

db = FAISS.from_documents(docs, embeddings)

results = db.similarity_search("Explain AI")
```

---

# 14 Retrieval Augmented Generation (RAG)

RAG = Retriever + LLM

```python
from langchain.chains import RetrievalQA

retriever = db.as_retriever()

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)

response = qa_chain.run("What is covered in the document?")
print(response)
```

---

# 15 Agents in LangChain

Agents can:

* Decide what tool to use
* Perform multi-step reasoning
* Use calculators, search APIs, etc.

Example:

```python
from langchain.agents import initialize_agent, load_tools
from langchain.agents import AgentType

tools = load_tools(["llm-math"], llm=llm)

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION
)

agent.run("What is 25 * 16?")
```

#  Most Asked LangChain Interview Questions (With 2-Minute Answers)

---

## 1 What is LangChain?

**Answer (2 min):**
**LangChain** is an open-source framework used to build applications powered by Large Language Models (LLMs). It provides tools to manage prompts, chains, memory, document retrieval (RAG), agents, and external tool integrations.

Instead of calling an LLM API directly, LangChain helps structure workflows like:

User Input  Prompt  LLM  Post-processing  Output

It is widely used for chatbots, document Q&A systems, AI agents, and automation pipelines.

---

## 2 Why was LangChain introduced?

**Answer (2 min):**
Raw LLM APIs only take text input and return text output. Building full applications requires:

* Prompt engineering
* Memory handling
* External data retrieval
* Multi-step reasoning
* Tool usage

LangChain was introduced to provide a structured framework for managing these components efficiently, making LLM application development scalable and modular.

---

## 3 What is a Chain in LangChain?

**Answer (2 min):**
A **Chain** is a sequence of components executed in order.

Example:

* Prompt Template
* LLM
* Output Parser

Chains allow us to connect multiple steps. The output of one step becomes the input of the next.

There are different types:

* LLMChain
* SequentialChain
* SimpleSequentialChain

Chains are deterministic and follow a predefined workflow.

---

## 4 What is an Agent in LangChain?

**Answer (2 min):**
An **Agent** is a dynamic decision-making system.

Unlike chains (fixed flow), agents:

* Decide which tool to use
* Perform multi-step reasoning
* Call APIs or calculators dynamically

Agents use reasoning frameworks like ReAct (Reason + Act). They are used in AI assistants that need flexible behavior.

---

## 5 Difference Between Chain and Agent?

**Answer (2 min):**

| Feature    | Chain            | Agent                 |
| ---------- | ---------------- | --------------------- |
| Flow       | Predefined       | Dynamic               |
| Tool usage | Fixed            | Decides automatically |
| Reasoning  | Limited          | Multi-step reasoning  |
| Use case   | Simple workflows | AI assistants         |

Chains are predictable; Agents are intelligent and adaptive.

---

## 6 What is RAG in LangChain?

**Answer (2 min):**
RAG stands for **Retrieval-Augmented Generation**.

Instead of relying only on LLMs training data, RAG:

1. Retrieves relevant documents from a vector database.
2. Passes them to the LLM.
3. Generates a context-aware answer.

Steps:

* Load documents
* Split text
* Create embeddings
* Store in vector DB (like FAISS)
* Retrieve relevant chunks
* Generate answer

RAG improves accuracy and reduces hallucinations.

---

## 7 What are Embeddings in LangChain?

**Answer (2 min):**
Embeddings convert text into numerical vectors.

These vectors capture semantic meaning, allowing:

* Similarity search
* Document retrieval
* Clustering

In LangChain, embeddings are used with vector stores to enable RAG systems.

---

## 8 What is a Vector Store?

**Answer (2 min):**
A Vector Store stores embeddings and allows similarity search.

Common vector databases:

* FAISS
* Chroma

When a user asks a query:

* Query  Convert to embedding
* Search similar vectors
* Retrieve relevant documents

This powers document Q&A systems.

---

## 9 Why is Text Splitting Important?

**Answer (2 min):**
LLMs have token limits. Large documents must be split into smaller chunks.

Benefits:

* Avoids token overflow
* Improves retrieval accuracy
* Better semantic matching
* Faster processing

LangChain provides `RecursiveCharacterTextSplitter` for intelligent chunking.

Memory is important for:

* Chatbots
